```python
import csv
import pandas as pd

from common import *
```

```python
NAME = "Statement-Information_1"
```

```python
original = pd.read_csv(f"input/{NAME}.csv")
original
```

```python
producer = original.iloc[:, 2].tolist()
producer = [to_array(p) for p in producer]
```

```python
original.loc[2]
```

```python
consumer = [[] for _ in range(len(producer))]
for i, p in enumerate(producer):
    if p == None: 
        continue

    for _p in p:
        consumer[int(_p)-1].append(str(i+1))
```

```python
consumer = [to_string(c) for c in consumer]
```

```python
df = original.copy()
df.insert(3, "Consumer statement", consumer)
df
```

```python
COLUMNS = ['Procedure', 'Statement', 'producer statement', 'consumer statement',
                'cost', 'result size', 'Fragment']
df.columns = COLUMNS
```

```python
df.to_csv(f"input/{NAME}.csv", index=False)
```

```python
dfs = [[] for i in range(10)]
for index, row in df.iterrows():
    r = row.tolist()
    dfs[r[0]-1].append(r)
```

```python
for i, df in enumerate(dfs):
    d = pd.DataFrame(df, columns=COLUMNS)
    d.to_csv(f"input/{NAME}/Procedure-{i+1}.csv", index=False)
```
